```python
def get_buyer_seller_info(p):
    title = p['title']
    features = p.get('features','')
    desc = p.get('description','')
    if len(desc) < 5 and len(desc) < len(features):
        desc = features
    listPrice = float(p['list_price'].removeprefix('$').replace(',','')) 
    highPrice = float(p['highest_price'].removeprefix('$').replace(',',''))
    seller_target = listPrice
    buyer_target = float(p['lowest_price'].removeprefix('$').replace(',',''))
    category = p['category']
    return title, category, listPrice, {'buyer_item_description': desc, 'buyer_target_price':buyer_target}, {'seller_item_description':desc, 'seller_target_price':seller_target}

def get_buyer_seller_prompt(data_record):
    title, category, list_price, buyer_dict, seller_dict = get_buyer_seller_info(data_record)
    buyer_item_description = buyer_dict['buyer_item_description']
    buyer_item_description = f"""- Category: {category}
- Description:
{buyer_item_description}
"""

    seller_item_description = seller_dict['seller_item_description']
    seller_item_description = f"""- Category: {category}
- Description:
{seller_item_description}
"""
    return title, list_price, buyer_item_description, seller_item_description
```

```python
import json
import os

dir='./raw_data'
processed_data = []
for file in sorted(os.listdir(dir)):
    with open(os.path.join(dir,file)) as f:
        products = json.load(f)
        for p in products:
            title, list_price, buyer_item_description, seller_item_description = get_buyer_seller_prompt(p)
            # print("===")
            # print(buyer_initial_prompt)
            # print(seller_initial_prompt)
            processed_data.append({'title':title, 'list_price':list_price, 'buyer_item_description':buyer_item_description, 'seller_item_description':seller_item_description})

print(f"Length of dataset {len(processed_data)}")
with open("processed_data.json", "w") as file:
    json.dump(processed_data, file, indent=4)
```
