```python
import json
import os
from pprint import pprint

dir='./amazon/raw_data'

products = []
for file in sorted(os.listdir(dir)):
# file = sorted(os.listdir(dir))[0]
    with open(os.path.join(dir,file)) as f:
        products += json.load(f)
```

```python
def process_product(p):
    title = p['title']
    features = p.get('features','')
    desc = p.get('description','')
    highPrice = float(p['highest_price'].removeprefix('$').replace(',',''))
    highPriceDate = p['highest_price_date']
    lowPrice = float(p['lowest_price'].removeprefix('$').replace(',',''))
    lowPriceDate = p['lowest_price_date']
    midPrice = (highPrice + lowPrice) / 2
    category = p['category']
    buyer_item_description = f"""- Category: {category}
- Description:
{desc}
- Features:
{features}
"""

    seller_item_description = f"""- Category: {category}
- Description:
{desc}
- Features:
{features}
"""
    high_price_dict = {
        'highest_price': highPrice,
        'highest_price_date': highPriceDate
    }
    low_price_dict = {
        'lowest_price': lowPrice,
        'lowest_price_date': lowPriceDate
    }
    return title, midPrice, high_price_dict, low_price_dict, buyer_item_description, seller_item_description
```

```python
# get processed dataset
processed_data = []
for product in products:
    title, midPrice, high_price_dict, low_price_dict, buyer_item_description, seller_item_description = process_product(product)

    processed_data.append(
        {'title':title, 
         'mid_price':midPrice, 
         'buyer_item_description':buyer_item_description, 'seller_item_description':seller_item_description, 
         'highest_price_info':high_price_dict, 
         'lowest_price_info': low_price_dict})
```

```python
# I have 1402+930=2332 products in total
# For training, I will use ~80% of the data, which is 1875 products,
# For validation, I will use ~8.5% of the data, which is 200 products,
# For testing, I will use ~11.5% of the data, which is 257 products.

training_data = amazon_data[:750] + craigslist_data[:1125]
validation_data = amazon_data[750:850] + craigslist_data[1125:1225]
testing_data = amazon_data[850:] + craigslist_data[1225:]

print(f"Training data: {len(training_data)} products")
print(f"Validation data: {len(validation_data)} products")
print(f"Testing data: {len(testing_data)} products")

# save the data
with open("./training_data.json", "w") as file:
    json.dump(training_data, file, indent=4)
with open("./validation_data.json", "w") as file:
    json.dump(validation_data, file, indent=4)
with open("./testing_data.json", "w") as file:
    json.dump(testing_data, file, indent=4)
```

```python
import random

def sample_private_values(low_price, high_price, gain_from_trade=True, random_seed=None):
    """
    Sample private cost for the seller and private value for the buyer based on historical price data.

    There are two modes:
    1. Gain from Trade Mode: Ensures buyer value > seller cost, creating a zone of possible agreement (ZOPA).
    2. No Gain from Trade Mode: Ensures buyer value < seller cost, eliminating ZOPA.

    Parameters:
    - low_price (float): The historical lowest price of the item.
    - high_price (float): The historical highest price of the item.
    - gain_from_trade (bool): Whether the mode is "gain from trade" (True) or "no gain from trade" (False).
    - random_seed (int, optional): Seed for reproducible random sampling.

    Returns:
    - seller_cost (float): The private cost for the seller.
    - buyer_value (float): The private value for the buyer.

    Illustration of Ranges:
    -----------------------
    **Seller Cost (Consistent Across Modes):**
    Seller cost is sampled uniformly within:
        [0.6 * low_price, 0.95 * low_price]
    This range reflects realistic seller costs, typically below the historical lowest price.

    Example:
    If low_price = 100:
        Seller cost range = [60, 95]

    **Buyer Value (Dependent on Mode):**
    1. Gain from Trade Mode:
        Buyer value is sampled uniformly within:
            [seller_cost + 0.01, high_price + 0.2 * (high_price - low_price)]
        This ensures buyer value is always greater than seller cost, creating a ZOPA.

        Example:
        If low_price = 100, high_price = 200, and seller_cost = 80:
            Buyer value range = [80.01, 220]

    2. No Gain from Trade Mode:
        Buyer value is sampled uniformly within:
            [0.5 * low_price, seller_cost - 0.01]
        This ensures buyer value is always less than seller cost, eliminating ZOPA.

        Example:
        If low_price = 100 and seller_cost = 80:
            Buyer value range = [50, 79.99]

    Notes:
    - The ranges are designed to ensure realistic negotiation dynamics based on historical price data.
    - The random seed allows for reproducibility in experiments.
    """
    if random_seed is not None:
        random.seed(random_seed)
    
    if gain_from_trade:
        seller_cost = random.uniform(0.6 * low_price, 0.9 * low_price)
        buyer_value = random.uniform(seller_cost + 0.01, high_price + 0.1 * (high_price - low_price))
    else:
        seller_cost = random.uniform(0.6 * low_price, 0.9 * low_price)
        buyer_value = random.uniform(0.4 * low_price, seller_cost - 0.01)
    
    return round(seller_cost, 2), round(buyer_value, 2)
```

```python
# example usage, apply this to all products in the validation data
for product in validation_data:
    mid_price = product['mid_price']
    low_price = product['lowest_price_info']['lowest_price']
    high_price = product['highest_price_info']['highest_price']
    
    # sample private values for seller and buyer
    seller_cost, buyer_value = sample_private_values(low_price, high_price, gain_from_trade=True, random_seed=42)
    
    # add to the product
    product['seller_private_cost'] = seller_cost
    product['buyer_private_value'] = buyer_value

    # print the prices
    print("========================================")
    print(f"Product: {product['title']}")
    print(f"Mid Price: {mid_price}, Low Price: {low_price}, High Price: {high_price}")
    print(f"Seller Cost: {seller_cost}, Buyer Value: {buyer_value}")
```

```python
import numpy as np
from typing import List, Dict, Any, Optional


def sample_private_values_for_dataset(
    dataset: List[Dict[str, Any]],
    gain_from_trade: bool = True,
    random_seed: Optional[int] = None,
) -> List[Dict[str, Any]]:
    """
    Sample the seller's private cost and the buyer's private value for every listing in
    *dataset* using a deterministic NumPy random matrix.

    A 2‑column matrix ``R`` of independent draws from :math:`\text{Uniform}(0,1)` is
    produced first:

    * **Shape:** ``(N, 2)`` where ``N = len(dataset)``.
    * **Seeding:** ``numpy.random.seed(random_seed)`` if *random_seed* is not ``None``.
    * **Usage:**
        * ``R[i, 0]`` → seller‑cost draw for listing *i*.
        * ``R[i, 1]`` → buyer‑value draw for listing *i*.

    The raw draws are then linearly mapped into economically meaningful intervals while
    preserving the boundaries used in the original implementation.

    **Seller cost (all modes)**
        ``seller_cost_i = 0.6·low_i  +  R[i,0] · (0.9·low_i  − 0.6·low_i)``
        ∈ ``[0.6 × lowest_price, 0.9 × lowest_price]``

    **Buyer value**
        *Gain from trade* ``(gain_from_trade=True)``
            ``buyer_value_i = (seller_cost_i+0.01) + R[i,1] · (hi_i + 0.1·Δ_i − (seller_cost_i+0.01))``
            where ``hi_i = highest_price`` and ``Δ_i = hi_i − low_i``.
            Range: ``[seller_cost+0.01, highest_price + 0.1·(highest_price − lowest_price)]``.  
            This ensures buyers are willing to pay at least 1 ¢ more than the seller’s cost.

        *No gain from trade* ``(gain_from_trade=False)``
            ``buyer_value_i = 0.4·low_i  + R[i,1] · ((seller_cost_i−0.01) − 0.4·low_i)``
            Range: ``[0.4 × lowest_price, seller_cost − 0.01]``.  
            This enforces an efficiency loss by making the buyer’s valuation below the
            seller’s cost.

    The function **mutates** each product dictionary by adding two rounded entries:

    * ``'seller_private_cost'`` — float with 2‑decimal precision.
    * ``'buyer_private_value'`` — float with 2‑decimal precision.

    Parameters
    ----------
    dataset : list[dict]
        Each product must hold
        ``product['lowest_price_info']['lowest_price']`` and
        ``product['highest_price_info']['highest_price']``.
    gain_from_trade : bool, default=True
        Toggle between *gain* and *no‑gain* regimes described above.
    random_seed : int | None, default=None
        Random seed for full reproducibility; skipped when ``None``.

    Returns
    -------
    list[dict]
        The same list with the additional keys per listing.
    """

    # ── RNG initialisation ──────────────────────────────────────────────────────
    if random_seed is not None:
        np.random.seed(random_seed)

    n = len(dataset)
    random_draws = np.random.rand(n, 2)  # column‑0: seller, column‑1: buyer
    # print(random_draws)
    # ── Sampling loop ───────────────────────────────────────────────────────────
    for i, product in enumerate(dataset):
        low_price = product["lowest_price_info"]["lowest_price"]
        high_price = product["highest_price_info"]["highest_price"]

        # Seller cost mapping
        seller_min = 0.6 * low_price
        seller_max = 0.9 * low_price
        seller_cost = seller_min + random_draws[i, 0] * (seller_max - seller_min)

        # Buyer value mapping
        if gain_from_trade:
            buyer_min = seller_cost + 0.01
            buyer_max = high_price + 0.1 * (high_price - low_price)
        else:
            buyer_min = 0.4 * low_price
            buyer_max = seller_cost - 0.01

        # Guard against numerical issues where max == min
        if buyer_max <= buyer_min:
            buyer_max = buyer_min + 0.01

        buyer_value = buyer_min + random_draws[i, 1] * (buyer_max - buyer_min)

        # Attach rounded results back to product
        product["seller_private_cost"] = round(seller_cost, 2)
        product["buyer_private_value"] = round(buyer_value, 2)

    return dataset
```

```python
updated_validation_data = sample_private_values_for_dataset(validation_data, gain_from_trade=False, random_seed=30)

# Print the updated dataset
for product in updated_validation_data:  # Print first 5 products for brevity
    print("========================================")
    print(f"Product: {product['title']}")
    print(f"Low Price: {product['lowest_price_info']['lowest_price']}, High Price: {product['highest_price_info']['highest_price']}")
    print(f"Seller Cost: {product['seller_private_cost']}, Buyer Value: {product['buyer_private_value']}")
```
